```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covid_deaths = pd.read_csv('Datasets/covid19_deaths_dataset.csv')
covid_deaths.head()
```

```python
covid_deaths.dtypes
```

```python
covid_deaths.describe()
```

```python
covid_deaths.shape
```

```python
covid_deaths.drop(['Lat','Long'], axis=1,inplace=True)
covid_deaths.head()
```

```python
covid19_deaths_agg = covid_deaths.groupby(by='Country/Region').sum()
covid19_deaths_agg.head()
```

```python
covid19_deaths_agg.shape
```

```python
covid19_deaths_agg.loc['China'].plot()
covid19_deaths_agg.loc['Spain'].plot()
covid19_deaths_agg.loc['Italy'].plot()
covid19_deaths_agg.loc['Afghanistan'].plot()
covid19_deaths_agg.loc['Korea, South'].plot()
plt.legend()
```
